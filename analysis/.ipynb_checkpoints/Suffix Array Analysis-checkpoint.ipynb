# Suffix Array Algorithms Analysis


This notebook summarizes analysis for certain algorithms that build the [suffix array](
https://en.wikipedia.org/wiki/Suffix_array) data structure, classified according to their complexities.

1. **Brute Force** - *O(n^2)*
2. **Radix Sort** - *O(n log^2 n)*
3. **Counting Sort** - *O(n log n)*
4. **DC3 Algorithm** - *O(kn)*

Conclusion summary:
- Brute Force special case is best for all (letters == 1).
- Brute Force average complexity is inversely proportional to the increase of letters if (letters >= 2).
- Brute Force is **best** for small cases (length <= 100) in terms of average complexity.
- DC3 is **best** for medium cases (100 <= length <= 1000) with (2 <= letters < 10).
- Brute Force is **best** for medium cases (100 <= length <= 1000) with (letters >= 10).
- DC3 is **best** for all large test cases (length > 1000) with (letters >= 2).
- Counting Sort has the same best case and worst case *O(n log n)* complexity.
- Radix Sort beats Counting Sort's average case complexity.

```python
from progvar.strings.utils import random_text
from time import time

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
mpl.style.use('ggplot')

def test(trials=10, text_length=10000, letters=5, exclude=['default']):
    
    # increment length every trial
    lengths = range(text_length // trials, text_length + 1, text_length // trials)
    trial_text = [random_text(length=length, letters=letters) for length in lengths]
    
    def perform_algorithm(algorithm):
        
        trials = []
        suffix_arrays = []

        # profile time
        time_elapsed = 0
        total_text_length = 0
        total_letters = set()


        for text in trial_text:
            # get median for small test cases to avoid anomalies
            timer = []
            for i in xrange(10 if text_length <= 3000 else 1):
                time_start = time()
                result = SuffixArray(text, algo=algorithm)
                timer.append(time() - time_start)
            
            timer.sort()

            # collect time statistics
            
            delta_time = np.median(timer)
            time_elapsed += delta_time

            # collect text statistics
            total_text_length += len(text)
            total_letters.update(list(text))

            # append to results
            trials.append(delta_time * 1000)
            suffix_arrays.append(result)


        time_per_trial = time_elapsed / len(trial_text)
        statistics = {
            'algorithm': algorithm,
            '# of trials': len(trial_text),
            'total time (ms)': time_elapsed * 1000,
            'average time per trial': time_per_trial * 1000,
            'letters': len(total_letters),
            'length': total_text_length / len(trial_text),
        }
        return trials, statistics, suffix_arrays
    
    algorithms_used = []
    all_trials = []
    all_statistics = []
    all_suffix_arrays = []
    
    for algorithm in SuffixArray.algorithms:
        
        if algorithm in exclude:
            continue
            
        trials, statistics, suffix_array = perform_algorithm(algorithm)
        
        algorithms_used.append(algorithm)
        all_trials.append(trials)
        all_statistics.append(statistics)
        all_suffix_arrays.append(suffix_array)
    
    n = len(all_trials)
    for algo, sa in zip(algorithms_used, all_suffix_arrays):
        for algo2, sa2 in zip(algorithms_used, all_suffix_arrays):
            if algo < algo2 and sa != sa2:
                print "WARNING! %s and %s have different suffix arrays!" % (algo, algo2)
    
    algo_trial = [{} for i in xrange(len(all_trials[0]))]
    for algo, trial in zip(algorithms_used, all_trials):
        for j in xrange(len(trial)):
            algo_trial[j][algo] = trial[j]
            
    #plt.show()
    
    all_statistics.sort(key=lambda stat: stat['total time (ms)'])
    from IPython.display import display, HTML
    display(pd.DataFrame(all_statistics, columns=('algorithm', 'total time (ms)')))
    
    df = pd.DataFrame(algo_trial, columns=algorithms_used, index=lengths)
    df.plot(kind='area', stacked=False, color=['r', 'g', 'k', 'w'])
    plt.show()


def perform_tests(n):
    letter_counts = [1, 2, 5, 10, 26, 50]
    for letters in letter_counts:
        print letters, 'letters:'
        test(trials=10, text_length=n, letters=letters)
```

# # Small Test Cases (n <= 100)

```python
perform_tests(n=100)
```

## Medium Test Cases (100 <= n <= 1000)

```python
perform_tests(n=1000)
```

## Large Test Cases (1000 <= n <= 10000)

```python
perform_tests(n=10000)
```

## Extra Large Test Case (10000 <= n <= 100000)

```python
perform_tests(n=100000)
```
